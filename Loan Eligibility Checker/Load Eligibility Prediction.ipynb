```python
x_train,x_test,y_train,y_test = train_test_split(independent,dependent, test_size = 0.2,random_state = 42)
```

 SVM Model

```python
svc_params = {   
   "kernel" : ["rbf","sigmoid","linear"],   
   "C"      : [100,150],     
   "gamma"   : ["scale", "auto"]
}
svc_classifier = GridSearchCV(SVC(),svc_params, refit = True)  
svc_classifier.fit(x_train,y_train) 
print(svc_classifier.best_params_)
```

Decision Tree

```python
#Decision Tree Model

dt_params = {
    "criterion"    : ["gini", "entropy", "log_loss"],
    "splitter"     : ["best", "random"],
    "max_features" : ["sqrt", "log2"]
}

dt_classifier = GridSearchCV(DecisionTreeClassifier(),dt_params, refit = True)  
dt_classifier.fit(x_train,y_train) 
print(dt_classifier.best_params_)
```

```python
#Predicting
y_dt_pred = dt_classifier.predict(x_test)

#finding Confustion Matrix
dt_confusion_mtx = confusion_matrix(y_test,y_dt_pred)
print(dt_confusion_mtx)

#finding Classification Report
dt_classification_rpt = classification_report(y_test,y_dt_pred)
print(dt_classification_rpt)
```

Random Forest Model

```python
#Random Forest Model

rf_params = {
    "criterion"    : ["gini", "entropy", "log_loss"],
    "n_estimators" : [50,100,250,500,1000],
    "max_features" : ["sqrt", "log2",None]
}

rf_classifier = GridSearchCV(RandomForestClassifier(),rf_params, refit = True)  
rf_classifier.fit(x_train,y_train) 
print(rf_classifier.best_params_)
```

```python
#Predicting
y_rf_pred = rf_classifier.predict(x_test)

#finding Confustion Matrix
rf_confusion_mtx = confusion_matrix(y_test,y_rf_pred)
print(rf_confusion_mtx)

#finding Classification Report
rf_classification_rpt = classification_report(y_test,y_rf_pred)
print(rf_classification_rpt)
```

KNN Model

```python
#Knn Model

knn_params = {    
    "n_neighbors" : [5,10,100,150,200],
    "algorithm"   : ["auto", "ball_tree", "kd_tree", "brute"],
    "metric" : ["minkowski"]    
}

knn_classifier = GridSearchCV(KNeighborsRegressor(),knn_params, refit = True)  
knn_classifier.fit(x_train,y_train) 
print(knn_classifier.best_params_)
```
